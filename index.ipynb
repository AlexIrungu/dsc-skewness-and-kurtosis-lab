You'll generate two datasets and measure/visualize and compare their skew and kurtosis in this lab.

```python
# Import required libraries
import numpy as np
import matplotlib.pyplot as plt

from scipy.stats import kurtosis, skew
```

## Take 1
* Generate a random normal variable `x_random` in NumPy with 10,000 values. Set the mean value to 0 and the standard deviation to 2.
* Plot a histogram of the data, set bins to `auto` (default). 
* Calculate the skewness and kurtosis for this data distribution using the SciPy functions.
* Record your observations about the calculated values and the shape of the data. 

```python
# Setting a seed for consistent results
np.random.seed(42)

x_random = np.random.normal(loc=0, scale=2, size=10000)

# Plot a histogram of the data
plt.hist(x_random, bins='auto')
plt.title('Histogram of Randomly Generated Data')
plt.show()

# Calculate the skewness and kurtosis
skewness = skew(x_random)
kurtosis_value = kurtosis(x_random)

print(f"Skewness: {skewness}")
print(f"Kurtosis: {kurtosis_value}")

# With seed set to 42, and if you wrote your code the *exact* way we have it in the solution branch
# you should get these results:
# Skewness = 0.001963697766356243
# kurtosis = 0.026479272360444117
```

```python
# Your observations here 
# The skewness value is close to 0, indicating the data distribution is symmetrical.
# The kurtosis value is also close to 0, suggesting the data distribution has a similar peakiness or flatness to the normal distribution.
```

## Take 2

Let's generate another distribution 

```python
x = np.linspace( -5, 5, 10000 )
y = 1./(np.sqrt(2.*np.pi)) * np.exp( -.5*(x)**2  )  # normal distribution
```

* Plot a histogram for data $y$, and set bins to auto (default).
* Calculate the skewness and kurtosis for this data distribution using the SciPy functions.
* Record your observations about the calculated values and the shape of the data.

```python
# Your code here:
# Plot a histogram for data y, and set bins to auto (default).
plt.hist(y, bins='auto')
plt.title('Histogram of Normal Distribution')
plt.show()

# Calculate the skewness and kurtosis
skewness_y = skew(y)
kurtosis_y = kurtosis(y)

print(f"Skewness: {skewness_y}")
print(f"Kurtosis: {kurtosis_y}")

# Skewness = 1.109511549276228
# kurtosis = -0.31039027765889804
```

```python
# Your observations here 

# Given y represents a normal distribution, we expect the skewness to be close to 0,
# indicating a symmetrical distribution. The kurtosis should also be close to 0,
# suggesting a similar peakiness or flatness to the normal distribution.
```
